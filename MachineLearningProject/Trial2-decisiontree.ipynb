```python
#convert Yes values to 1 and No values to 0
new_ibm_df = new_ibm_df[(new_ibm_df["Attrition"] == "Yes") | (new_ibm_df["Attrition"] == "No")]
mapping_dictionary = {"Attrition": { "Yes": 1, "No": 0 } }
new_ibm_df = new_ibm_df.replace(mapping_dictionary)
new_ibm_df.head()
```
